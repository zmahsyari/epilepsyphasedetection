```python
import pandas as pd
import numpy as np
import itertools
```

```python
df = pd.read_csv("E:/Thesis/Data/Data Jadi/sz7.txt")
df.head()
```

```python
df = df.rename(columns={'0': 'ECG'})
df = df.drop(['Unnamed: 0'], axis=1)
df.head()
```

```python
ms = np.arange(0.00,len(df)*0.005,0.005)
df['second'] = ms.tolist()
```

```python
df['time'] = pd.to_datetime(df['second'], unit='s').dt.strftime('%H:%M:%S')
```
